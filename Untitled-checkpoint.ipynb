```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from scipy.stats import norm
df = pd.read_csv("Billionaires Statistics Dataset.csv", index_col="personName") # Indexing specific column
df[["age", "finalWorth", "source", "countryOfCitizenship"]].head(20)   # Take the first 100 rows of the data with specified columns
```

```python
countries = ['United States']
filt = df['countryOfCitizenship'].isin(countries)
df.loc[filt, ['rank','countryOfCitizenship']].head(20)
#Filtered top 20 billioners from US
```

```python
filt_net_worth = df['finalWorth']>90000
filt_net_worth.head(20)
df.loc[filt_net_worth, ['countryOfCitizenship', 'finalWorth']]
#Filtered Billionaires who have more than 90 billion
```

```python
df.sort_values(by=['finalWorth', 'age'], ascending=[True, False], inplace=True)
df[['countryOfCitizenship', 'age']].head(20)
```

```python
print(int(df.age.median()), "Years old")  # AVG Age of top 30 billioners
print(int(df.age.min()), 'Years old') #Youngest Billioner age
print(df.finalWorth.max(), "USD" )   #Max Networth
```

```python
#Top 7 richest people and their networth
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(20, 10))
names = df['personName'].head(7)
net_worth = df["finalWorth"].head(7)
ax.bar(names, net_worth, color = 'b', width = 0.5)
ax.set_xlabel('Names', fontsize = 25, color = 'r')
ax.set_ylabel('Networth', fontsize = 25, color = 'r')
ax.set_title('Persons and their networth')

plt.show()
```

```python
#Top 10 richest people and their share
import matplotlib.pyplot as plt
labels = df['personName'].head(10)
sizes = df["finalWorth"].head(10)
fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct='%.0f%%',
       textprops={'size': 'smaller'}, radius=1)
plt.show()
```
